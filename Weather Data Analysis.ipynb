```python
#step2:
#Create or load the data - the weather dataset
#Assume dataset with columns: Date, Temperature, Humidity, Rainfall

#pd.date_range is a function in the pandas library that generates a sequence of dates. It's incredibly useful for creating a range of dates for time series analysis or any other date-related operations. 
#Here's a basic example of how to use it:
#generate a synthetic data

#syntax:        
#pandas.date_range(start=None, end=None, periods=None, freq=None, tz=None, normalize=False, name=None, inclusive='both', *, unit=None, **kwargs)

#Parameters:
#1)Start Parameter - String or datetime- like, optional
#left bound for generating dates

#2) end Parameter - String or datetime- like, optional
#right bound for generating dates

#3) Frequency - https://pandas.pydata.org/docs/reference/api/pandas.date_range.html
#yyyy-mm-dd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data ={'date': pd.date_range(start='2024-01-01', end='2024-12-31', freq = 'D'), #daily dates for one year
              'Temperature':np.random.randint(15,40, size = 366), #(start stop size)
               'Humidity': np.random.randint(30, 90, size = 366), #Random Humidity percentages
               'Rainfall': np.random.uniform(0, 50, size = 366) #uniform means slowly increase

            }
data

#now create a DataFrame

df = pd.DataFrame(data)
df.head() # displayed first 5 rows
print(df['date'].dtype)

#Here the date is in datetime format so no need to convert
```

```python
#Step 3: Add a 'Month' Column
#We extract the month from the Date column to group data by months.

df['Month'] = df['date'].dt.month_name()
df.sample()
```

```python
#Step 4: Calculate Monthly Averages
#Group the data by the Month column and calculate the average of Temperature, Humidity, and Rainfall.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
monthly_averages = df.groupby('Month')[['Temperature','Humidity','Rainfall']].mean()
monthly_averages


monthly_averages = df.groupby('Month')[['Temperature', 'Humidity', 'Rainfall']].mean()
#object[?]
```

```python
#As the above date is not in the order of months we can sort in order(like Jan, Feb, March....) - this is optional

month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

monthly_averages = monthly_averages.reindex(month_order)
monthly_averages
```

```python
#Step 5: Identify the Month with the Highest Rainfall
#We find the month with the highest average rainfall using .idxmax()
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

highest_rainfall_month = monthly_averages['Rainfall'].idxmax()
print(f"The month with the highest average rainfall is: {highest_rainfall_month}")
```

```python
#Step 6: Plot Daily Temperature Trends
#Create a line plot to visualize temperature trends across the year.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))  # Set the figure size
plt.plot(df['date'], df['Temperature'], color='green', label='Temperature')

# Add labels and title
plt.title('Daily Temperature Trends in 2024', fontsize=16)
plt.xlabel('Date', fontsize=25)
plt.ylabel('Temperature (°C)', fontsize=20)

# Rotate x-axis labels for better visibility
#plt.xticks(rotation=45)

# Add grid and legend
plt.grid(True, linestyle='--', alpha=0.6) 
# plt.legend()

# Show the plot
# plt.tight_layout()
plt.show()
```

```python
#Step 7: Plot Monthly Average Temperature
#Create a bar chart to visualize monthly average temperatures.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize = (12,6)) #first set the figure size
plt.bar(monthly_averages.index, monthly_averages['Temperature'], color = 'skyblue')
plt.title('Monthly Average Temperature', fontsize=12)
plt.xlabel('Months', fontsize=12)
plt.ylabel('Average Temperature (°C)', fontsize=12)
plt.show()
```

```python
#Step 8: Plot Monthly Average Rainfall
#Create a bar chart to visualize monthly average rainfall for comparison.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.figure(figsize = (12,6))
plt.bar(monthly_averages.index, monthly_averages['Rainfall'], color = 'olive')
plt.title("Monthly Average Rainfall", fontsize = 15)
plt.xlabel("Months", fontsize = 15)
plt.ylabel("Average Rainfall", fontsize = 15)
plt.show()
```

```python
#1. Bar Plot - A Bar plot is a simpe and effective way to visualize categorical data with rectangular bars. 
#The length of each bar is proportional to the value it represents.

#Example:
#Imagine you are running a fruit shop and want to track how many apples, oranges, lemons, and limes you sold today. 
#You can visualize the sales with bar plot.

#Apples: 10, Oranges: 15, Lemons: 5, Limes:12

#The bar plot will easily compares the sales figures across different fruits

#For bar plot we need categorical data

import matplotlib.pyplot as plt

categories = ['Apples', 'Oranges', 'Lemons', 'Limes']
values = [10, 15, 5, 12]

plt.bar(categories, values)
plt.title('Fruit Sales')
plt.xlabel('Fruits')
plt.ylabel('Quantity Sold')
plt.show()
```

```python
#2. Line Plot

#A Line plot is used to visualize the trend of a variable over time or some other continuous measure

#Lets say you are monitoring the temperature over a day. you can use line plot to track how the temperature changes at different times.
#A smooth curve shows whether the temperature rises or falls

#TIME - 6AM, 9AM, 12 PM, 3 PM, 6 PM, 9 PM - Temperature - 18°C, 22°C, 27°C, 24°C, 20°C, 17°C

import matplotlib.pyplot as plt

time = ['6 AM', '9 AM', '12 PM', '3 PM', '6 PM', '9 PM'] 
temperature = ['18°C', '22°C', '27°C', '24°C', '20°C', '17°C']
plt.plot(time, temperature, color = 'green', marker = 'o')
plt.title("Temperature over the day")
plt.xlabel("Time")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
#3, Violin plot - A violin plot is a method of plotting numeric data to show its distributiin,
#providing deeper intensity than a box plot by showing density

#Imagine you have a test scores for three different classes and you want to understand how the scores are spread across the students.
#The violin plot shows the distribution, helping you see which class has higher variability in scores

import matplotlib.pyplot as plt
import numpy as np
data = [np.random.normal(0, std, 100) for std in range(1,4)]
plt.violinplot(data, showmeans = True)
plt.title("Violin plot of test scores")
plt.show()

# The violin plot is best for showing data distribution, density, and comparisons between multiple categories, with richer detail than a box plot.
```

```python
#4. Histogram - It is a type of bar chart that represents the distribution of numerical data by showing the frequency of data
#within certain ranges

#Imaging you are collecting data on the ages of people in your neighbourhood, and you want to see how the ages are distributed 
#A histogram can help you visualixe the frequency of each group.


import matplotlib.pyplot as plt

ages = [12, 23, 31, 34, 43, 54, 35, 38, 45, 56, 60, 59, 62, 65, 43, 69, 70]

plt.hist(ages, bins = 5, edgecolor = 'black')
plt.title("Age distribution in a Neighbourhood")
plt.xlabel("Ages")
plt.ylabel("Frequency")
plt.show()
```

```python
#5.Scatterplot

# Imagine you are studying the relationship between the amount of time spent studying and the scores achieved in an exam. 
# A scatter plot shows whether there is a positive or negative correlation between these variables.


import matplotlib.pyplot as plt

study_hours= [2, 4, 6, 8, 10]

scores = [50,55,70,80,95]
plt.scatter(study_hours, scores, color = 'green', marker = 'o')
plt.title("Study hours vs Exam scores")
plt.xlabel("Time spent")
plt.ylabel("Exam scores")
plt.show()             
```

```python
#6. Pie chart - A Pie chart represents data in slices where each slice corresponds to categories contribution to the whole.
#You want to show the market share of different companies in Tech industry. Apie chart helpd to show the proprtion if each company's market share.

import matplotlib.pyplot as plt

labels = ['Facebook', 'Apple', 'Netflix', 'Google']
sizes = [35, 37, 40, 38]
explode = (0, 0, 0.1, 0)
plt.pie(sizes, labels = labels, explode = explode, autopct = '%1.1f%%', shadow = True, startangle = 90)
plt.title("Market share of the companies")
plt.show()
```
